```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
import pandas as pd
```

```python
data = pd.read_csv('creditcard.csv')
```

```python
data.head(5)
```

```python
train = data.iloc[:190000,:].copy()
test = data.iloc[190000:,].copy()
```

```python
model = LogisticRegression()
```

```python
model.fit(X=train.loc[:,['Amount']],y=train.Class)
```

```python
def logistic_transformation(x):
    return 1/(1+np.exp(-x))
```

```python
model.coef_
```

```python
model.intercept_
```

```python
test['prob'] = logistic_transformation(0.00019734*test.Amount -6.23920183)
```

```python
test[(test['Class'] == 1)].head(5)
```

```python
set(model.predict(test.loc[:,['Amount']]))

len(test)
```

```python
display(model.predict_proba(test.loc[:,['Amount']])[:,-1])  ##标签为1的概率

len(model.predict_proba(test.loc[:,['Amount']])[:,-1])
```
